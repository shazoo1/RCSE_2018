### - Delta rule (Widrow-Hoff rule)

- The **delta rule** is a gradient descent learning rule for updating the weights of the inputs to artificial neurons in single-layer neural network.
- The goal is to **minimize** the error between the actual outputs and the target outputs in the training data.
- For each (input/output) training pair, the delta rule **determines** the direction you need to adjust $W_ij$ to reduce the error for that training pair.
- **Derivatives** are used for teaching

#### Gradient descent method

Find the steepest way down the slope from where you are, and take a step in that direction.

Alorithm:
1. Define learning rate $ \alpha$ (0 < $ \alpha  < 1$) and $ E_{min} $
2. Initialize the weights with some small values
3. Take input pattern and calculate the output vector
4. Modify the weights and bias according delta rule
5. Do steps 3-4 until $ E $ > $ E_{min} $
