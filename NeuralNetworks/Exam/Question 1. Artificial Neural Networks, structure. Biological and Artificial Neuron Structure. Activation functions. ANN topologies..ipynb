## 1. Artificaial Neural Networks, structure. Biological and Artificial Neuron Structure. Activation functions. ANN topologies.

### Artificial intellegience (AI)
is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and other animals.

Design of ANN is motivated by their **similarity** to successfully working **biological systems**, which consist of very simple but numerous nerve cells that work massively in parallel and have the capabilty to learn.

Benefits:
- Self organization and learning capabilities
- Implicitly programmed
- Capability to generalize and associate data
- Fault tolerant against internal errors and noise, redundancy and sharing of responsibilities.
- Applicable if rules are unknown or complicated
- Memory is distributed and content addressable

Drawbacks:
- ANN is a blackbox;
- it is difficult to recognize the point of fault

### Artificial Neural Network (ANN)
is an extremely simplified model of the human's brain. Transform input into the best fit outputs



### Biological structure




Main components:
- Cell - main core on a neuron. Life center
- Dendrities - receive messages from other cells
- Axon - pass output to other cells
- Synapse - A point of connection between an axon of an other neuron and a dendrite

### Artificial Neuron Structure



Main components:
- x - inputs
- '*' - multipliers
- W - weights of synapses
- T - bies
- f - activation function
- y - output

### Activation function

Activation function maps the resulting values into the desired range.

Most used activation functions:
- linear
- logistic
- tanh
- tan-sigmoid
- binary threshold
- sign
- relu
- sigmoid

### Topologies

Some well-known ANN topologies:
- Single layer ANN(aka Perceptron)
- Multilayer ANN (MNN)
- Hopfield
- Convolutional ANN (CNN)
- Generative adversarial network (GAN)
