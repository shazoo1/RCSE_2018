## Roseblattâ€™s single layer perceptron. Training.

- The perceptron is an algorithm for supervised classification of an input into one of several possible non-binary outputs.
- It is a type of linear classifier.
- Was invented in 1957 by Frank Rosenblatt as a machine for image recongition

### Learning algorithm

1. Initialize the weights and the threshold. Weights may be initialized with 0 or a random small value.
2. Take input pattern $x$ from $X$ and calculate output vector $y$ from $Y$.
3. If $y_i$ = $t_j$ then $w_{ij}$ will not be changed
4. If $y_i$ != $t_j$ then $w_{ij} (t+1) = w_{ij} (t) + \alpha x_i t_j$ 
5. Do steps 2-4 until $y_i$ = $t_j$ for the whole training set (if it is possible)

- It was quickly proved that perceptrons could not be trained to recognize many classes of patterns.
- It is linear classifier. For example, it is impossible for these classes of network to learn an XOR function.
