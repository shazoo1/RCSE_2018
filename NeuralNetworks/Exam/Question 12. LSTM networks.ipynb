## Long-short-term memory

Long-short-term memory (LSTM) block or network is a **simple recurrent neural network** which can be used as a building component or block (of hidden layers) for an eventually bigger recurrent neural network. The expression **long-short-term** refers to the fact that LSTM is a model for the **short-tiem memory** which can last for a **long-period of time**.

An LSTM block is composed of four main components:

1. A cell
2. An input gate
3. An output gate
4. A forget gate

### Cell
A cell is responsible for "remembering" values over arbitrary time intervals; hence the word "memory" in LSTM

### Gate

Each of the gate has its own parameters (i.g. weights and biases from possible other units the LSTM block).

Their output is multiplied with the output of the cell or the input to the LSTM to partially allow or deny information to flow into or out of the memory.

#### - Input gate
The input gate controls the extend to which a new value flows into the memory


#### - Forget gate 
Controls the extent to which a value remains in memory

#### - Output gate
Controls the extent to which the value in memory is used to compute the output activation of the LSTM block


In some implementations, the input and forget gates are merged into a single gate. The motivation for combining them is that the time to forget is when a new value worth remembering becomes available (**context**)

### Applications

1. Speech recognition
2. Text translation
3. Times series forecasting ...
