## Question 1. Hopfiend neural network. Training

Hopfield NN - is recurrent neural networks, i.e. **includess the network output in the following comutation steps**

- Invited by John Hopfield in 1982.
- **Content-addressable memory** with binary threshold nodes (-1,1 or 0,1)
- Weight matrix is symmetric **to be stable (convergence to the attractor)**

### Hopfield NN as associative memory

### Algorithm

1. Take noise pattern $y$
2. Implement iterations

3. Do again until we reach stable state (attractor)

### Modes

#### Syncronous mode

All neurons process information simultaneously (converges to 1 stable pattern or cycle with the lenght of 2).

#### Asyncronous mode

Always converges to 1 stable pattern - neurons process in information one by one. Neuron element selects randomly from the remaining list

$$ y_j(t+1)=sign(S_i(t))$$
$$ S_j(t)=Y(t)W_j$$
