- Unsupervices training
    - The training set **only** consist of **input patterns**.
    - The neural network adjusts its own weights so that **similar inputs cause similar outputs**. The network identifies the patterns and diffirences in the inputs **without any external assistance**.
- Clustering - cluster analysis or clustering is the task of **grouping a set of objects** in such a way that **objects in the same group** (called a cluster) are **more similar to each other** than to those in other groups (clusters).

Goals:
- Data compressionss
- Detection the outliers (anomalies)
- Taxonomy (construction the objects hierarchy)

Clustering approaches:
- Hierarchical clustering
- K-means (centroid based)
- EM clustering (discribution based)
- DBSCAN (density based)

## Competititve learning

* We only ask **which neuron is active  at the moment**
* We are not interested in the exact output of the neuron but to know **which neuron provides the output**

**Conpetitive learning** is a form of unsupervies learning in ANN, in which nodes **compete for the right to respond**

### Self-organizing maps (SOM) or Kohonen

- It is a type of ANN that is trained using **unsupervised learning** to produce a low-dimentional (typically, two-dimentional), discretized representation of the input space of the training samples, called a map.
- Self-organizing maps are different from other artificial neural networks in the sense that they use a neighborhood function to preserve the **topological** properties of the input space.
- The model was first described as an ANN by Finnish professor **Teuvo Kohonen**
- Uses competitive learning
