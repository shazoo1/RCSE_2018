## The structure of multilayer perceptrons. Backpropagation of error.

### Multilayer perceptron
A multilayer perceptron (MLP) is a feed forward artificial neural network model that maps sets of input data onto a set of appropriate outputs

- **Consists of multiple layers** (input, output, one of several hidden layers) of nodes in a directed graph, with each layer fully connected to the next one.
- Neurons with a **nonlinear activation function**
- Utilizes a supervised learning technique called **backpropagation of error**

Calculation the output Y for input vector X



Features:
- Activation function is not a threshold, usually a sigmoid func is used
- Function approximator, not limited to linear problems
- Information flows in one direction, the outputs of one layer act as inputs to the next layer

- A single layer network can only find a linear discriminant function.
- MLP with one hidden layer can approximate arbitrarly closely every continuous function
- MLP with two hidden layers can classify sets of any form

### - Backpropagation algorithm

- Designed by D.Rumelhard, G.Hinton, R.Williams (1986)
- Most common method of obtaining the weights in the multilayer perceptron
- A form of supervised training
- The basic backpropagation algorithm is based on minimizing the error of the network using the derivatives of the error function
- Backpropagation of error generalizes the delta rule

Basic steps:

1. Forward propagation of a training pattern's input
2. Backward propagation of the output's error to generatee the deltas of all output and hidden neurons
