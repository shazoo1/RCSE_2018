## Question 2. Types of learning. Supervised and unsupervised learning. Fundamentals of learning and training samples. Error measurement.

### - Types of learning
Knowledge can be acquired through experience or education by perceiving, discovering, or learning.  


Sources of knowledge:
- From expert( aka expert system) - applicable if well-defined rules accessible with precise input data
- From data (supervised and unsupervised)

#### Supervised learning

The system is supplied with a set of training examples.
Examples consists of inputs and corressponding expected outputs.
The system is required to discover the relation or mapping between them.

- Supplies an ANN with inputs and the correct outputs
- A error vector can be estimated for a certain input
- Response of the ANN is measured. The weights modified according error vector to reduce the diffirence between the received result and the desired output.

#### Unsupervised learning
The system is supplied with a set consisting only of inputs.
The system is required to recognise the relation by itself, so an appropriate output is discovered.

- The training set only consists of input patterns
- The ANN adjusts its own weights, so that similar inputs cause similar outputs.
- The ANN has to identify the patterns and the diffirence in the input without any external assistance.

### - Fundamentals of learning and training samples

- The **weights** in a ANN are the most important factor in determining its function
- **Training** is the act of presenting the network with some sample data and modifying the weights to better approximate the desired function
- A **training set** is a set of **training patterns**, which we use to train our neural net.
- A **training pattern** is an input vector $p$ with the components $x_1$, $x_2$,..,$x_n$ whose desired output is known.
- The set of training patterns is called $P$. It contains a finite number of ordered pairs $(p, t)$ of training patterns with corresponding desired output.
- **Teaching input** - an input always refers to a specific training pattern $p$ and contained in the set $P$ of the training patterns.
- **Error vector** - for several output neurons $\sigma_1$, $\sigma_2$,..., $\sigma_n$ the diffirence between output vector and teaching input under a training input $p$ is referred to as error vector.



- **Epoch** - single presentation of the entire data to the neural network. In the other words, an amount of data for one full iteration of training process. Typically many epochs are required to train the neural network
- **Iteration** - the process of providing the network with an single input and updating the network's weights

### Error measurement

Let $ \sigma $ be the output and $O$ to be the set of output neurons.

The specific error $Err_p$ is based on a single training sample.


The total error $Err$ is based on all training samples.


Generally, the training process is stopped when the user in front of the learning computer "thinks" the error is small enough.
