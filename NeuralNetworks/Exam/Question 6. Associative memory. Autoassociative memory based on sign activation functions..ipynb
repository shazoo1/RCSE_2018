## Associative memory. Autoassociative memory based on sign activation functions.

- ### Associative memory

Associative memory (computer science) - a data-storage device in which a **location** is identified by its **informational content** rather than by names, addresses, or relative positions, and from which the data may be retrieved. This memory enables one to **retrieve** a piece of **data** from only a tiny sample of itself.

Associative memory (phychology) - recalling a previously experienced item by thinking of something that is linked with it, thus invoking the association.

- **Authoassociative memories**  are capable of retrieving a piece of data upon presentation of only partial information from that piece of data.
- **Heteroassociative memories** can recall an associated piece of datum from one category upon presentation of data from another category.
